```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt

dataset = pd.read_csv('emails.csv')
dataset
```

```python
#drop duplicates
dataset.drop_duplicates(inplace = True)
#check null values
print(pd.DataFrame(dataset.isnull().sum()))
```

```python
#create test and train data

X = dataset.text
y = dataset.spam

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.50)

#CountVectorizer() function was used in order to change each email into a vector counting 
#the number of times each word occurs.

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
train_text = cv.fit_transform(X_train.values)
spam = y_train.values
```

```python
#fit the naive bayes classifier
from sklearn.naive_bayes import MultinomialNB
classifier = MultinomialNB(alpha=1.0, class_prior=None, fit_prior=True)
classifier.fit(train_text,spam)

# Predicting the Test set results and checking the accuracy

#Transform the test data in order to fit into the model
test_text = cv.transform(X_test.values)
y_pred = classifier.predict(test_text)

from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred) 
```
